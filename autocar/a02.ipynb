```python
import cv2 #4.3버전
from pop import Util

Util.enable_imshow()

```

```python
cam  = Util.gstrmer()
```

```python
cap = cv2.VideoCapture(cam, cv2.CAP_GSTREAMER)
if not cap.isOpened():
    print("Not found camera")
try:
    for _ in range(12000):
        ret, frame = cap.read()
        if not ret:
            break
        img = cv2.Canny(frame, 100, 200)
        cv2.imshow("soda", img)
except KeyboardInterrupt:
        pass
finally:
    cap.release()
#cv2.destroyAllwindows():윈도우 종료한는건데 주피터에서는 의미없음
```
